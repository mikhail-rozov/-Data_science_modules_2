##### 1. В чём различие между зависимыми и независимыми выборками?

Зависимость или независимость выборок - больше вопрос организации данных, то, как мы представляем выборки.
Независимые выборки - выборки, которые не влияют друг на друга. В зависимых выборках данные одной выборки оказывают влияние на другую выборку.  
Разберем на примере нашего учебного датасета.

```python
import pandas as pd
pd.options.display.max_columns = 100
```

```python
df = pd.read_csv('./training_project_data.csv')
```

```python
df
```

Возьмём две выборки:

```python
df.head(10)
```

```python
df.tail(10)
```

В данном случае эти выборки являются независимыми, т.е. данные первой выборки не оказывают влияния на данные второй выборки, потому что принадлежат разным людям, никак друг с другом не связанным, кроме того факта, что они являются клиентами одного банка.

Теперь возьмём другие выборки:

```python
df.iloc[:10, :12]
```

```python
df.iloc[:10, 12:]
```

Эти выборки являются зависимыми, т.к. соответствуют одним и тем же людям, и данные, находящиеся в первой выборке, оказывают влияние на данные второй выборки.

##### 2. Когда применяются параметрические статистические критерии, а когда — их непараметрические аналоги?

Параметрические критерии применяются тогда, когда наша выборка имеет определенный тип распределения (в основном, нормальный), т.к. такие критерии основаны на свойствах такого распределения. Эти критерии обладают большей мощностью относительно непараметрических.  
То есть, если мы определили, например, с помощью критерия Шапиро-Уилка, что наша выборка имеет нормальное распределение, то лучше отдать предпочтение именно параметрическому критерию (например, критерию Стьюдента). Если мы не можем определить тип распределения выборки, то нужно использовать непараметрические критерии (например, критерий Манна-Уитни).
